```julia
using Plots
```

```julia
function sigma(x)
    if -1<=x<=-0.5
        y = 1
    elseif -0.5<=x<=0.5
        y = 6(0.5-x)^5-15(0.5-x)^4+10(0.5-x)^3
    elseif 0.5<=x<=1
        y = 0
    end
    return y
end
        
```

```julia
function g1_eps(x,eps)
    
    y = 1-sigma(x)*exp(-(1+x)/eps)-sigma(-x)*exp(-(1-x)/eps)
    return y
end
```

```julia
epsilon = 0.01
x = -1:0.01:1
y = g1_eps.(x,epsilon)
plot(x,y)
```
